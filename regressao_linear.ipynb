```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_saude = pd.read_csv("/content/plano_saude.csv")
```

```python
df_saude
```

```python
x_plano_saude = df_saude["idade"].values
y_plano_saude = df_saude["custo"].values
```

```python
np.corrcoef(x_plano_saude,y_plano_saude)
```

```python
x_plano_saude = x_plano_saude.reshape(-1,1)
```

```python
print(x_plano_saude,"\n\n",y_plano_saude)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressao_plano_saude = LinearRegression()
```

```python
regressao_plano_saude.fit(x_plano_saude,y_plano_saude)
```

```python
previsoes = regressao_plano_saude.predict(x_plano_saude)
```

```python
previsoes
```

```python
from sklearn.metrics import accuracy_score
```

```python
grafico = px.scatter(x = x_plano_saude.ravel(), y = y_plano_saude)
grafico.add_scatter(x = x_plano_saude.ravel(),y = previsoes, name = "regressão" )
grafico.show()
```

```python
regressao_plano_saude.score(x_plano_saude,y_plano_saude)
```

```python
x_casa = base_casas["sqft_living"].values.reshape(-1, 1)
y_casa = base_casas["price"].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_casa_treinamento,x_casa_teste, y_casa_treinamento, y_casa_teste = train_test_split(x_casa,y_casa, test_size = 0.3)
```

```python
print(x_casa_treinamento.shape, x_casa_teste.shape, y_casa_treinamento.shape,y_casa_teste.shape)
```

```python
regressor_casa = LinearRegression()
```

```python
y_casa_treinamento
```

```python
regressor_casa.fit(x_casa_treinamento, y_casa_treinamento)
```

```python
regressor_casa.score(x_casa_treinamento, y_casa_treinamento)
```

```python
regressor_casa.score(x_casa_teste, y_casa_teste)
```

```python
previsoes = regressor_casa.predict(x_casa_treinamento)
```

```python
print(len(previsoes), len(x_casa_treinamento.ravel()))
```

```python
grafico = px.scatter(x = x_casa_treinamento.ravel(), y = previsoes.ravel())
grafico.show()
```

```python
grafico1 = px.scatter(x = x_casa_treinamento.ravel(), y = y_casa_treinamento.ravel())
grafico2 = px.line(x = x_casa_treinamento.ravel(), y= previsoes.ravel())
grafico2.data[0].line.color = 'red'
grafico3 = go.Figure(data = grafico1.data + grafico2.data)
grafico3.show()
```
